```python
# import standscaler and variance inflation factor
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = []
for i in range(scaled.shape[1]):
    X = variance_inflation_factor(scaled,i)
    vif.append(X)
```

```python
# import all model building lib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso, Ridge, LassoCV, RidgeCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score, GridSearchCV
import xgboost as xgb
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled,y,test_size=0.30,random_state=105)
```

```python
# linear regression model
LR = LinearRegression()
LR.fit(x_train,y_train)
y_pred = LR.predict(x_test)
print('R2_score:',r2_score(y_test,y_pred))
print('Training Score:',LR.score(x_train,y_train))
print('Testing Score :',LR.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,y_pred))
print('mean squared error :', mean_squared_error(y_test,y_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
lassocv = LassoCV(alphas=None,max_iter=20)
lassocv.fit(x_train,y_train)
alpha = lassocv.alpha_
alpha
```

```python
# Lasso regression model 
lasso = Lasso(alpha,max_iter=20)
lasso.fit(x_train,y_train)
lasso_pred = lasso.predict(x_test)
print('R2_score:',r2_score(y_test,lasso_pred))
print('Training Score:',lasso.score(x_train,y_train))
print('Testing Score :',lasso.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,lasso_pred))
print('mean squared error :', mean_squared_error(y_test,lasso_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,lasso_pred)))
```

```python
ridgeCV = RidgeCV(alphas=np.arange(0.001,0.1,0.01))
ridgeCV.fit(x_train,y_train)
ridgeCV.alpha_
```

```python
# Ridge regression model 
ridge = Ridge(ridgeCV.alpha_)
ridge.fit(x_train,y_train)
ridge_pred = ridge.predict(x_test)
print('R2_score:',r2_score(y_test,ridge_pred))
print('Training Score:',ridge.score(x_train,y_train))
print('Testing Score :',ridge.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,ridge_pred))
print('mean squared error :', mean_squared_error(y_test,ridge_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,ridge_pred)))
```

```python
# GradientBoosting regression model 
gb = GradientBoostingRegressor()
gb.fit(x_train,y_train)
gb_pred = gb.predict(x_test)
print('R2_score:',r2_score(y_test,gb_pred))
print('Training Score:',gb.score(x_train,y_train))
print('Testing Score :',gb.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,gb_pred))
print('mean squared error :', mean_squared_error(y_test,gb_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,gb_pred)))
```

```python
# Randamforest Regression model 
RF = RandomForestRegressor()
RF.fit(x_train,y_train)
RF_pred = RF.predict(x_test)
print('R2_score:',r2_score(y_test,RF_pred))
print('Training Score:',RF.score(x_train,y_train))
print('Testing Score :',RF.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,RF_pred))
print('mean squared error :', mean_squared_error(y_test,RF_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,RF_pred)))
```

```python
# AdaBoost Regression model
ada = AdaBoostRegressor()
ada.fit(x_train,y_train)
ada_pred = ada.predict(x_test)
print('R2_score:',r2_score(y_test,ada_pred))
print('Training Score:',ada.score(x_train,y_train))
print('Testing Score :',ada.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,ada_pred))
print('mean squared error :', mean_squared_error(y_test,ada_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,ada_pred)))
```

```python
# Bagging Regression model 
bg = BaggingRegressor()
bg.fit(x_train,y_train)
bg_pred = bg.predict(x_test)
print('R2_score:',r2_score(y_test,y_pred))
print('Training Score:',bg.score(x_train,y_train))
print('Testing Score :',bg.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,bg_pred))
print('mean squared error :', mean_squared_error(y_test,bg_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,bg_pred)))
```

```python
# KNeighbors Regression model
knn = KNeighborsRegressor()
knn.fit(x_train,y_train)
knn_pred = knn.predict(x_test)
print('R2_score:',r2_score(y_test,knn_pred))
print('Training Score:',knn.score(x_train,y_train))
print('Testing Score :',knn.score(x_test,y_test))
print('mean absolute error :', mean_absolute_error(y_test,knn_pred))
print('mean squared error :', mean_squared_error(y_test,knn_pred))
print('Root means squared error :', np.sqrt(mean_squared_error(y_test,knn_pred)))
```
